```python
from graphius import Graph
```

```python
adjacency_list = {
    'w': ['r', 'v', 'x', 'z'],
    'x': ['w', 'y', 'z'],
    'y': ['u', 'v', 'x'],
    'z': ['w', 'x'],
    'r': ['s', 't', 'w'],
    's': ['r', 'u', 'v'],
    't': ['r', 'u'],
    'u': ['t', 's', 'y'],
    'v': ['s', 'w', 'y'],
}

g = Graph(adjacency_list)
visited, visual = g.bfs('s', visualized=True)
print(visited)
g.print_path('s', 'z')

g.visualize()
```

```python
adjacency_list = {
    's': ['w', 'z'],
    't': ['u', 'v'],
    'u': ['t', 'v'],
    'v': ['s', 'w'],
    'w': ['x'],
    'x': ['z'],
    'y': ['x'],
    'z': ['w', 'y']
}

g = Graph(adjacency_list, directed=True)
g.dfs(visualized=True)
length = len(g.names)*2
for i in range(1, length+1):
    for vertex in g.vertices.values():
        if vertex.d == i:
            print(f'({vertex.name}', end=' ')
            break
        elif vertex.f == i:
            print(f'{vertex.name})', end=' ')
g.visualize()
```
